```python
!pip install transformers torch
from transformers import DistilBertTokenizer, DistilBertForQuestionAnswering
import torch
```

```python
tokenizer = DistilBertTokenizer.from_pretrained(model_path)
model = DistilBertForQuestionAnswering.from_pretrained(model_path)
print("✅ Model and tokenizer loaded successfully!")
```

```python
def ask_faq(question, context):
    inputs = tokenizer.encode_plus(question, context, return_tensors="pt")
    outputs = model(**inputs)
    start = torch.argmax(outputs.start_logits)
    end = torch.argmax(outputs.end_logits) + 1
    tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0][start:end])
    return tokenizer.convert_tokens_to_string(tokens)
```

```python
context = """Health care / Medical Cyber-Physical Systems (MCPS)

What it is & architecture
MCPS (aka healthcare CPS or Internet of Medical Things) combine wearable/implantable sensors, bedside medical devices, hospital equipment, local gateways/edge nodes, cloud analytics, and clinician interfaces. They form closed-loop control (monitor → analyze → act) for diagnosis, monitoring, and automated therapy. Typical layers: sensing (WBANs), edge processing, secure communication, cloud/analytics, clinician dashboard / actuation.

Representative applications
⦁	Continuous patient monitoring (ECG, SpO₂, glucose), closed-loop insulin pumps / infusion pumps, remote ICU monitoring, smart operating rooms, automated drug delivery, and AI-assisted imaging devices.

Key technical & clinical challenges
Safety & real-time guarantees: control loops (e.g., insulin delivery) must meet strict timing and correctness requirements. Security & privacy: devices and wireless links are targets for attacks; patient data confidentiality is critical and device integrity must be assured to avoid life-threatening manipulation. Interoperability: many vendors, proprietary protocols, legacy devices in hospitals. Resource constraints: many medical sensors are battery-powered, with limited compute and bandwidth. Regulation & validation: medical devices require rigorous testing, certification, and clinical validation.

Design best practices
⦁	Defense-in-depth security (device attestation, encryption, secure boot), fail-safe modes and human overrides, real-time OS / formal verification where needed, strong access control and audit trails, use of edge processing to reduce latency and preserve privacy.

Trends
⦁	More edge AI for low-latency inference; blockchain/ledger ideas for auditability; converged standards for interoperability and secure device lifecycle management."""
```

```python
ask_faq("what is the architecture?", context)
```
